### Calculadora simple 

```python
num1 = float(input("Ingrese el primer número: "))
num2 = float(input("Ingrese el segundo número: "))

operacion = input("Ingrese la operación (+, -, *, /): ")

if operacion == "+":
  resultado = num1 + num2
elif operacion == "-":
  resultado = num1 - num2
elif operacion == "*":
  resultado = num1 * num2
elif operacion == "/":
  if num2 == 0:
    print("Error: No se puede dividir por cero.")
  else:
    resultado = num1 / num2
else:
  print("Operación inválida.")

print("El resultado de la operación es:", resultado)
```

### Contador de Números Pares e Impares

```python
numeros = range(1, 20)

pares = 0
impares = 0

for numero in numeros:

  if numero % 2 == 0:
    pares += 1
  else:
    impares += 1

print("Cantidad de números pares:", pares)
print("Cantidad de números impares:", impares)
```

### Validación de Contraseña

```python
contrasena_correcta = "python123"

continuar = True

while continuar:
  contrasena_usuario = input("Ingrese la contraseña: ")

  if contrasena_usuario == contrasena_correcta:
        print("Contraseña correcta!")
    continuar = False
  else:
        print("Contraseña incorrecta. Inténtelo de nuevo.")

```

### Identificador de Números Primos

```python
numero = int(input("Ingrese un número entero: "))

es_primo = True

if numero <= 1:
  es_primo = False

for i in range(2, numero):
    if numero % i == 0:
    es_primo = False
    break

if es_primo:
  print(f"{numero} es un número primo.")
else:
  print(f"{numero} no es un número primo.")
```

### Juego de Adivinanza

### Tabla de Multiplicar

```python
numero = int(input("Ingrese un número entero: "))

for multiplicador in range(1, 11):
    producto = numero * multiplicador
    print(f"{numero} x {multiplicador} = {producto}")
```
